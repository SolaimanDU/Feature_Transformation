```python
import pandas as pd
```

```python
df = pd.read_csv('Real estate.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x = df.drop(['Y house price of unit area'], axis = 1)
```

```python
x.columns
```

```python
y = df['Y house price of unit area']
```

```python
x.shape
```

# Min Max Scaler

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
new_x1 = mms.fit_transform(x)
x = pd.DataFrame(new_x1, columns = ['No', 'X1 transaction date', 'X2 house age',
       'X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.30, random_state = 1, shuffle = True)
```

```python
xtrain
```

```python
xtest
```

```python
scaled_test1 = mms.transform(xtest)
```

```python
scaled_test1
```
